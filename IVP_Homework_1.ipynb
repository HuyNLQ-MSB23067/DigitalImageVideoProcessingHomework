## **L5_1: Image Subtraction**

```python
!curl -o live.jpg https://raw.githubusercontent.com/HuyNLQ-MSB23067/DigitalImageVideoProcessingHomework/main/HW1/live.jpg
!curl -o mask.jpg https://raw.githubusercontent.com/HuyNLQ-MSB23067/DigitalImageVideoProcessingHomework/main/HW1/mask.jpg
liveImgData = cv2.imread('live.jpg', cv2.IMREAD_UNCHANGED)
maskImgData = cv2.imread('mask.jpg', cv2.IMREAD_UNCHANGED)
```

```python
liveImg = cv2.normalize(liveImgData.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
maskImg = cv2.normalize(maskImgData.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
```

```python
diffImg = abs(maskImg - liveImg)
```

```python
gamma = 0.3
diffImg_contrasted = diffImg ** gamma
```

```python
plt.subplot(2, 2, 1)
plt.imshow(liveImg)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(maskImg)
plt.title('Mask Image')

plt.subplot(2, 2, 3)
plt.imshow(diffImg)
plt.title('Difference Image')

plt.subplot(2, 2, 4)
plt.imshow(diffImg_contrasted)
plt.title('Difference Image With Contrast = {}'.format(gamma))
```

## **L5_2: Image Comparison**

```python
!curl -o mask1.png https://raw.githubusercontent.com/HuyNLQ-MSB23067/DigitalImageVideoProcessingHomework/main/HW1/mask1.png
!curl -o mask2.png https://raw.githubusercontent.com/HuyNLQ-MSB23067/DigitalImageVideoProcessingHomework/main/HW1/mask2.png
mask1ImgData = cv2.imread('mask1.png', cv2.IMREAD_UNCHANGED)
mask2ImgData = cv2.imread('mask2.png', cv2.IMREAD_UNCHANGED)
```

```python
compareImg = abs(mask1Img - mask2Img);
```

```python
plt.subplot(1, 3, 1)
plt.imshow(mask1Img)
plt.title('Mask Image #1')

plt.subplot(1, 3, 2)
plt.imshow(mask2Img)
plt.title('Mask Image #2')

plt.subplot(1, 3, 3)
plt.imshow(compareImg)
plt.title('Difference Image')
```

## **L5_3: Image Averaging**

```python
!curl -o quadnight.jfif https://raw.githubusercontent.com/HuyNLQ-MSB23067/DigitalImageVideoProcessingHomework/main/HW1/quadnight.jfif
ImgData = cv2.imread('quadnight.jfif', cv2.IMREAD_UNCHANGED)
ImgData = cv2.cvtColor(ImgData, cv2.COLOR_BGR2GRAY)
```

```python
Img = cv2.normalize(ImgData.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
```

```python
nArray = [1, 2, 8, 16, 32, 64];
for i in range(0, len(nArray)):
  N = nArray[i]
  avgImg = np.zeros(Img.shape)

  for j in range(0, N):
    noisyImg = Img + 0.4 * Img.std() * np.random.random(Img.shape)
    avgImg = avgImg + noisyImg
  
  avgImg = avgImg / N;
  plt.subplot(2, np.size(nArray)/2, i + 1)
  plt.imshow(avgImg)
  plt.title('{} image(s)'.format(N))
```
